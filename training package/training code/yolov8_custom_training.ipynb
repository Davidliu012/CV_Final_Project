## Training
* Upload your training data to the folder(ex: yolov8_train_detect) under your Google Drive.
* The training data should include file: data.yaml and folders: train, val, test.
* The output model will be under the folder(ex: yolov8_train_detect/training_results/busdoor/weights/).
* We train two models for the final project.



```python
!yolo task=detect mode=train model=yolov8s.pt data=/content/drive/MyDrive/yolov8_train_detect/data.yaml epochs=20 imgsz=640 batch=8 project=/content/drive/MyDrive/yolov8_train_detect/training_results name=busdoor
```

```python
!yolo task=segment mode=train model=yolov8n-seg.pt data=/content/drive/MyDrive/yolov8_train_segment/data.yaml epochs=100 imgsz=640 batch=8 project=/content/drive/MyDrive/yolov8_train_segment/training_results name=busdoor
```
